```python
import torch
theta_numerator = torch.arange(0, 512, 2).float() 

print(theta_numerator.shape)
print(theta_numerator)
```

```python
theta = 1.0 / (10000.0 ** (theta_numerator / 512))
print(theta.shape)
print(theta)
```

```python
m =  torch.arange(50) # (seq_len, 1)
print(m.shape)
print(m)
```

```python
freqs = torch.outer(m, theta).float()
print(freqs.shape)
print(freqs)
```

```python
freqs_complex = torch.polar(torch.ones_like(freqs), freqs) # (seq_len, head_dim // 2)
print(freqs_complex.shape)
print(freqs_complex)
```

```python
freqs_complex = freqs_complex.unsqueeze(0).unsqueeze(2)
print(freqs_complex.shape)
print(freqs_complex)
```

```python
x = torch.randn(2, 3, 4)
x_out = x.view(-1)  # 将 x 展平为一维张量
print(x_out.shape)  # 输出: torch.Size([24])
print(x.shape)  # 输出: torch.Size([2, 3, 4])
# 恢复原始形状
x_out = x_out.reshape(*x.shape)
print(x_out.shape)  # 输出: torch.Size([2, 3, 4])
```
